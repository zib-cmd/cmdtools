```python
%load_ext autoreload
%autoreload 2
import cmdtools as cmd
import cmdtools.systems.diffusion as dif
import matplotlib.pyplot as plt
import numpy as np
```

```python
m=dif.TripleWell(nx=8, ny=8, beta=10)
plt.imshow(m.u)
Q=m.Q
p = cmd.analysis.pcca.PCCA(Q,3)
def plotchi(chi):
    plt.figure()
    plt.imshow(np.reshape(chi, np.shape(m.u)))
plotchi(p.chi[:,0])
plotchi(p.chi[:,1])
plotchi(p.chi[:,2])
```
